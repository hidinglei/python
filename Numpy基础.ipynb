```python
# 10. 拷贝与视图
a = np.arange(12)
b = a  # 完全不拷贝
print(b is a)  # True

c = a.view()  # 视图（浅拷贝）
print(c is a)  # False
print(c.base is a)  # True
c[4] = 999
print(a)

s = a[1:6]  # 切片是一种视图
s[:] = 888
print(a)

d = a.copy()  # 深拷贝
print(d is a)  # False
print(d.base is a)  # False
d[4] = 777
print(a)
```

```python
# 25. 随机漫步1：原生Python方法
import random
position = 0
walk = [position]
steps = 1000
for i in range(steps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)
print(walk)
```

```python
# 26. 随机漫步2：numpy
nsteps = 1000
draws = np.random.randint(0, 2, size=nsteps)
steps = np.where(draws > 0, 1, -1)
walk = steps.cumsum()
print(walk)
```

```python
# 27. 随机漫步3: 统计
print(walk.min(), walk.max())
(np.abs(walk) >= 30).argmax()
```

```python
# 28. 随机漫步4: 多次模拟
nwalks = 5000
nsteps = 1000
draws = np.random.randint(0, 2, size=(nwalks, nsteps))
steps = np.where(draws > 0, 1, -1)
walks = steps.cumsum(axis=1)
print(walks)
```

```python
# 29. 随机漫步5:多次模拟的统计
print(walks.max(), walks.min())
print((np.abs(walks) >= 30).any(axis=1).sum())
print((np.abs(walks) >= 30).argmax(axis=1).mean())
```
